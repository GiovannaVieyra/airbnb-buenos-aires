```python
import pandas as pd

listings_raw = pd.read_csv("../data_raw/listings.csv")
calendar_raw = pd.read_csv("../data_raw/calendar.csv")
reviews_raw = pd.read_csv("../data_raw/reviews.csv")

print("LISTINGS:")
display(listings_raw.head())

print("CALENDAR:")
display(calendar_raw.head())

print("REVIEWS:")
display(reviews_raw.head())
```

```python
cols_listings = [
    "id",
    "name",
    "host_id",
    "host_name",
    "neighbourhood",
    "latitude",
    "longitude",
    "room_type",
    "price",
    "minimum_nights",
    "number_of_reviews",
    "last_review",
    "reviews_per_month",
    "calculated_host_listings_count",
    "availability_365",
]

listings = listings_raw[cols_listings].copy()

listings.head()
```

```python
listings["price_numeric"] = (
    listings["price"]
    .astype(str)                     
    .str.replace("$", "", regex=False)   
    .str.replace(",", "", regex=False)  
    .str.strip()                          
    .astype(float)                        
)

listings[["price", "price_numeric"]].head()
```

```python
listings["last_review"] = pd.to_datetime(listings["last_review"], errors="coerce")

listings["reviews_per_month"] = listings["reviews_per_month"].fillna(0)

listings = listings.dropna(subset=["price_numeric", "latitude", "longitude"])

listings.info()
```

```python
listings["price_30_days"] = listings["price_numeric"] * 30
```

```python
q1 = listings["price_numeric"].quantile(0.33)
q2 = listings["price_numeric"].quantile(0.66)

def price_category(p):
    if p <= q1:
        return "low"
    elif p <= q2:
        return "medium"
    return "high"

listings["price_category"] = listings["price_numeric"].apply(price_category)

listings[["price_numeric", "price_category"]].head()
```

```python
listings.to_csv("../data_clean/listings_clean.csv", index=False, encoding="utf-8")
```

```python
calendar = calendar_raw[
    ["listing_id", "date", "available", "price", "adjusted_price", "minimum_nights", "maximum_nights"]
].copy()
```

```python
calendar["date"] = pd.to_datetime(calendar["date"], errors="coerce")
```

```python
calendar["available"] = calendar["available"].map({"t": 1, "f": 0})
```

```python
for col in ["price", "adjusted_price"]:
    calendar[col] = (
        calendar[col].astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    calendar[col] = pd.to_numeric(calendar[col], errors="coerce")
```

```python
calendar.to_csv("../data_clean/calendar_clean.csv", index=False, encoding="utf-8")
```

```python
reviews = reviews_raw[
    ["listing_id", "id", "date", "reviewer_id", "reviewer_name", "comments"]
].copy()

reviews.head()
```

```python
reviews["date"] = pd.to_datetime(reviews["date"], errors="coerce")
```

```python
reviews["year"] = reviews["date"].dt.year 
```

```python
reviews = reviews.dropna(subset=["date"])
```

```python
reviews.to_csv("../data_clean/reviews_clean.csv", index=False, encoding="utf-8")
```
